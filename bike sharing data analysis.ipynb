## Overview of the dataset

the bike sharing dataset contains the following fields:

- instant: record index
	- dteday : The date on which the data was observed - in this case, the data was collected daily; so there's one row per date.
	- season : A numerically encoded value indicating the season (1:spring, 2:summer, 3:fall, 4:winter)
	- yr : he year of the study in which the observation was made (the study took place over two years - year 0 represents 2011, and year 1 represents 2012)
	- mnth : The calendar month in which the observation was made (1:January ... 12:December)
	- holiday : A binary value indicating whether or not the observation was made on a public holiday)
	- weekday : The day of the week on which the observation was made (0:Sunday ... 6:Saturday)
	- workingday : A binary value indicating whether or not the day is a working day (not a weekend or holiday)
	+ weathersit : A categorical value indicating the weather situation
		- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
		- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
		- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
		- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
	- temp : Normalized temperature in Celsius.
	- atemp: Normalized apparent ("feels-like") temperature in Celsius.
	- hum: Normalized humidity level.
	- windspeed: Normalized wind speed.
	- casual: count of rentals recorded by casual users
	- registered: count of rentals recorded by registered users
	- cnt: total count of rental for both casual and registered users

```python
# importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# import the data

df_bikeSharing = pd.read_csv("data/daily-bike-sharing-data.csv")
```

```python
df_bikeSharing.info()
```

The data does not contain any null values thus no need for data cleaning.

```python
# adding a new column for year/month
df_bikeSharing['month_year'] = pd.to_datetime(df_bikeSharing['dteday']).dt.to_period('M')
```

```python
# adding a new column for day of the month
df_bikeSharing['day'] = pd.DatetimeIndex(df_bikeSharing['dteday']).day
```

```python
# statistical information on the data

df_bikeSharing.describe()
```

We can observe a higher standard deviation for the 3 columns (casual, registered, and cnt) suggesting a high variantion in the count of rentals.

```python
# rental distribution for the registered users

# Get the label column
label = df_bikeSharing['registered']


# Create a figure for 2 subplots (2 rows, 1 column)
fig, ax = plt.subplots(2, 1, figsize = (9,12))

# Plot the histogram   
ax[0].hist(label, bins=100)
ax[0].set_ylabel('Frequency')

# Add lines for the mean, median, and mode
ax[0].axvline(label.mean(), color='magenta', linestyle='dashed', linewidth=2)
ax[0].axvline(label.median(), color='cyan', linestyle='dashed', linewidth=2)

# Plot the boxplot   
ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Rentals')

# Add a title to the Figure
fig.suptitle('Rental Distribution')

# Show the figure
fig.show()
```

```python
# gathering the numeric and categorical features

numeric_features = ['temp', 'atemp', 'hum', 'windspeed']
categorical_features = ['season','mnth','month_year','holiday','weekday','workingday','weathersit', 'day']
```

```python
# distribution of each numeric feature

for col in numeric_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = df_bikeSharing[col]
    feature.hist(bins=100, ax = ax)
    ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
    ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
    ax.set_title(col)
plt.show()
```

```python
# distribution for each categorical feature

for col in categorical_features:
    counts = df_bikeSharing[col].value_counts().sort_index()
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    counts.plot.bar(ax = ax, color='steelblue')
    ax.set_title(col + ' counts')
    ax.set_xlabel(col) 
    ax.set_ylabel("Frequency")
plt.show()
```

```python
# side by side plot of the number of rentals for casual users and registered users for each of the categorical features

for col in categorical_features:
    X = df_bikeSharing[col].unique()
    casual = df_bikeSharing.groupby(col)['casual'].sum()
    registered = df_bikeSharing.groupby(col)['registered'].sum()

    X_axis = np.arange(len(X))

    plt.bar(X_axis - 0.2, casual, 0.4, label = 'casual')
    plt.bar(X_axis + 0.2, registered, 0.4, label = 'registered')

    plt.xticks(X_axis, X)
    plt.xlabel("Groups")
    plt.ylabel("Number of Students")
    plt.title("Number of Students in each group")
    plt.legend()
    plt.show()
```

```python
# Checking the relationship between the numerical features and the number of rentals for casual users

for col in numeric_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = df_bikeSharing[col]
    label = df_bikeSharing['casual']
    correlation = feature.corr(label)
    plt.scatter(x=feature, y=label)
    plt.xlabel(col)
    plt.ylabel('Bike Rentals')
    ax.set_title('rentals vs ' + col + '- correlation: ' + str(correlation))
plt.show()
```

```python
# Checking the relationship between the numerical features and the number of rentals for registered users

for col in numeric_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = df_bikeSharing[col]
    label = df_bikeSharing['registered']
    correlation = feature.corr(label)
    plt.scatter(x=feature, y=label)
    plt.xlabel(col)
    plt.ylabel('Bike Rentals')
    ax.set_title('rentals vs ' + col + '- correlation: ' + str(correlation))
plt.show()
```

```python
# Checking the relationship between the numerical features and the number of rentals for all users

for col in numeric_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = df_bikeSharing[col]
    label = df_bikeSharing['cnt']
    correlation = feature.corr(label)
    plt.scatter(x=feature, y=label)
    plt.xlabel(col)
    plt.ylabel('Bike Rentals')
    ax.set_title('rentals vs ' + col + '- correlation: ' + str(correlation))
plt.show()
```

```python
# Checking the relationship between the categorical features and the number of rentals for casual users

for col in categorical_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    df_bikeSharing.boxplot(column = 'casual', by = col, ax = ax)
    ax.set_title('Label by ' + col)
    ax.set_ylabel("Bike Rentals")
plt.show()
```
